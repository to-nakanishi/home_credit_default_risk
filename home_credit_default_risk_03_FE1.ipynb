```python
#1-1　異常値フラグ(Max/95％タイル値＞100)

outlier_cols = [
    'BU_AMT_CREDIT_MAX_OVERDUE_min', 'CC_SK_DPD_DEF_sum', 'POS_SK_DPD_DEF_sum',
    'CC_SK_DPD_DEF_mean', 'BU_AMT_CREDIT_MAX_OVERDUE_mean', 'POS_SK_DPD_sum',
    'CC_SK_DPD_sum', 'PREV_SELLERPLACE_AREA_min', 'POS_SK_DPD_DEF_mean',
    'BU_AMT_CREDIT_MAX_OVERDUE_std', 'BU_AMT_CREDIT_MAX_OVERDUE_max',
    'BU_AMT_CREDIT_MAX_OVERDUE_sum', 'CC_SK_DPD_mean', 'CC_SK_DPD_DEF_max',
    'CC_SK_DPD_DEF_std', 'PREV_SELLERPLACE_AREA_mean', 'PREV_SELLERPLACE_AREA_latest',
    'POS_SK_DPD_mean', 'PREV_DAYS_LAST_DUE_1ST_VERSION_min', 'BU_AMT_CREDIT_MAX_OVERDUE_latest',
    'PREV_SELLERPLACE_AREA_max', 'BU_AMT_ANNUITY_latest', 'BU_AMT_ANNUITY_std',
    'PREV_SELLERPLACE_AREA_std', 'BU_AMT_ANNUITY_sum', 'BU_AMT_ANNUITY_max',
    'BU_AMT_ANNUITY_min', 'POS_SK_DPD_DEF_std', 'PREV_SELLERPLACE_AREA_sum',
    'BU_AMT_ANNUITY_mean', 'POS_SK_DPD_std', 'POS_SK_DPD_DEF_max',
    'BU_AMT_CREDIT_SUM_min', 'BU_AMT_CREDIT_SUM_DEBT_min', 'CC_SK_DPD_max',
    'AMT_INCOME_TOTAL', 'CC_SK_DPD_std', 'CC_AMT_INST_MIN_REGULARITY_min',
    'POS_SK_DPD_max', 'BU_AMT_CREDIT_SUM_LIMIT_latest', 'BU_AMT_CREDIT_SUM_std',
    'BU_AMT_CREDIT_SUM_mean', 'CC_AMT_DRAWINGS_OTHER_CURRENT_mean', 'BU_AMT_CREDIT_SUM_sum',
    'INS_AMT_PAYMENT_min', 'CC_AMT_DRAWINGS_OTHER_CURRENT_sum', 'BU_BB_STATUS_NUM_MEAN_sum',
    'BU_AMT_CREDIT_SUM_LIMIT_mean', 'AMT_REQ_CREDIT_BUREAU_QRT', 'BU_AMT_CREDIT_SUM_DEBT_sum',
    'INS_AMT_INSTALMENT_min', 'PREV_AMT_DOWN_PAYMENT_min', 'BU_AMT_CREDIT_SUM_max',
    'CC_AMT_PAYMENT_CURRENT_min'
]

def apply_abnormal_flags(train_df, test_df, cols):
    # 1. 必要なカラムだけ抽出（存在しない場合は無視）
    target_cols = [c for c in cols if c in train_df.columns]

    # 2. フラグだけを格納する辞書
    train_new_flags = {}
    test_new_flags = {}

    for col in target_cols:
        # 計算時だけ float64 として扱う
        series_train = train_df[col].astype(float)
        q95_train = series_train.quantile(0.95)

        new_col_name = f'{col}_abnormal_MAX_frag'
        divisor = q95_train + 1e-9

        # 判定後、即座に int8 (メモリ最小) に変換
        train_new_flags[new_col_name] = (series_train / divisor > 100).astype('int8')
        test_new_flags[new_col_name] = (test_df[col].astype(float) / divisor > 100).astype('int8')

    # 3. 最後に一括結合（メモリ負荷を最小化）
    train_df = pd.concat([train_df, pd.DataFrame(train_new_flags, index=train_df.index)], axis=1)
    test_df = pd.concat([test_df, pd.DataFrame(test_new_flags, index=test_df.index)], axis=1)

    return train_df, test_df
print(len(outlier_cols))
```

```python
#3-1 FE(試走結果に基づくEDA(第1有力指標))　EXT_SOURCE_1～3の平均

def apply_ext_source_mean(df):
    ext_src_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
    # 3つのスコアの平均（欠損値を除いて計算）
    df['EXT_SOURCES_MEAN'] = df[ext_src_cols].mean(axis=1)

    return df
```

```python
#3-2 FE(試走結果に基づくEDA(第1有力指標))　EXT_SOURCE_1～3の平均(分母補正あり)

def apply_custom_ext_mean(df):

    # EXT_SOURCE_1～3の特殊補正平均
    # 分子：1〜3の合計（欠損は0扱い）
    # 分母：2 + (EXT_2が存在すれば1、なければ0)

    # 欠損を0に置換して合計を計算
    ext_sum = df[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].fillna(0).sum(axis=1)

    # 分母の計算: EXT_SOURCE_2が欠損していなければ1、していれば0を返す
    ext2_notnull = df['EXT_SOURCE_2'].notnull().astype(int)
    denominator = 2 + ext2_notnull

    # 特殊平均の算出
    df['EXT_SOURCES_CUSTOM_MEAN'] = ext_sum / denominator

    return df
```

```python
#3-3 FE(試走結果に基づくEDA(第1有力指標))　スコアごとのデフォルト率(EXT_SOURCE_1～3)

from sklearn.model_selection import KFold

def apply_ext_score_bin_te(train_df, test_df, target_col='TARGET', n_folds=5, smoothing=10):
    train_res = train_df.copy()
    test_res = test_df.copy()

    col = 'EXT_SOURCES_CUSTOM_MEAN'
    new_col = 'EXT_SOURCES_BIN_TARGET_RATE'

    bins = np.arange(0, 1.05, 0.05)

    train_res['ext_bin'] = pd.cut(train_res[col], bins=bins, include_lowest=True, labels=False)
    test_res['ext_bin'] = pd.cut(test_res[col], bins=bins, include_lowest=True, labels=False)

    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    global_mean = train_res[target_col].mean()

    oof_values = np.zeros(len(train_res), dtype=np.float64)

    for train_idx, val_idx in kf.split(train_res):
        df_t = train_res.iloc[train_idx]
        df_v = train_res.iloc[val_idx]

        agg = df_t.groupby('ext_bin')[target_col].agg(['count', 'mean'])
        counts = agg['count']
        means = agg['mean']
        smooth = (means * counts + global_mean * smoothing) / (counts + smoothing)

        oof_values[val_idx] = df_v['ext_bin'].map(smooth).values

    oof_values = np.where(np.isnan(oof_values), global_mean, oof_values)
    train_res[new_col] = oof_values

    agg = train_res.groupby('ext_bin')[target_col].agg(['count', 'mean'])
    smooth = (agg['mean'] * agg['count'] + global_mean * smoothing) / (agg['count'] + smoothing)
    test_values = test_res['ext_bin'].map(smooth).values.astype(np.float64)
    test_values = np.where(np.isnan(test_values), global_mean, test_values)
    test_res[new_col] = test_values

    train_res = train_res.drop(columns=['ext_bin'])
    test_res = test_res.drop(columns=['ext_bin'])

    return train_res, test_res
```

```python
#3-4(1)　FE(試走結果に基づくEDA(第1有力指標))　幾何平均(EXT_SOURCE_1～3)
def analyze_ext_nan_substitute(df):
    ext_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

    for col in ext_cols:
        print(f"--- {col} の調査 ---")

        # 1. 有効値があるデータのデフォルト率を0.05刻みで計算
        temp = df[[col, 'TARGET']].dropna(subset=[col]).copy()
        # ビン化（0.05刻み）
        bins = np.arange(0, 1.05, 0.05)
        temp['bin'] = pd.cut(temp[col], bins=bins, include_lowest=True)

        # スコア帯ごとのデフォルト率
        bin_stats = temp.groupby('bin')['TARGET'].mean()

        # 2. NaNデータのデフォルト率
        nan_default_rate = df[df[col].isnull()]['TARGET'].mean()

        # 3. 最もデフォルト率が近いスコア帯を特定
        # 差分の絶対値が最小のインデックスを取得
        closest_bin = (bin_stats - nan_default_rate).abs().idxmin()
        closest_bin_rate = bin_stats[closest_bin]

        # 指定の形式で表示
        print(f"NaNデータのデフォルト率: {nan_default_rate:.4f}")
        print(f"最もデフォルト率が近いスコア帯: {closest_bin}")
        print(f"最もデフォルト率が近いスコア帯のデフォルト率: {closest_bin_rate:.4f}")
        print(f"※推奨される補完用スコア値 (中央値): {closest_bin.mid:.4f}")
        print("\n")

# 実行（trainデータを使用）
analyze_ext_nan_substitute(train)
```

```python
#3-4(2)　FE(試走結果に基づくEDA(第1有力指標))　幾何平均(EXT_SOURCE_1～3)

def apply_ext_source_geometric_mean(train_df, test_df):
    """
    EXT_SOURCE_1～3の幾何平均（3乗根）を生成
    欠損時は、NaNデータのデフォルト率に基づいた特定スコアで補完
    """
    train_res = train_df.copy()
    test_res = test_df.copy()

    # 調査結果に基づく補完値の設定
    fill_values = {
        'EXT_SOURCE_1': 0.3750,
        'EXT_SOURCE_2': 0.5250,
        'EXT_SOURCE_3': 0.3750
    }

    ext_cols = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

    for col in ext_cols:
        fv = fill_values[col]
        # 欠損値を指定のスコアで補完した一時的なカラムを作成
        train_res[f'{col}_filled'] = train_res[col].fillna(fv)
        test_res[f'{col}_filled'] = test_res[col].fillna(fv)

    # 幾何平均の計算（3つの乗算の1/3乗）
    train_res['EXT_SOURCES_GEOM_MEAN'] = (
        train_res['EXT_SOURCE_1_filled'] * train_res['EXT_SOURCE_2_filled'] * train_res['EXT_SOURCE_3_filled']
    ) ** (1/3)

    test_res['EXT_SOURCES_GEOM_MEAN'] = (
        test_res['EXT_SOURCE_1_filled'] * test_res['EXT_SOURCE_2_filled'] * test_res['EXT_SOURCE_3_filled']
    ) ** (1/3)

    # 作業用カラムの削除
    drop_cols = [f'{col}_filled' for col in ext_cols]
    train_res = train_res.drop(columns=drop_cols)
    test_res = test_res.drop(columns=drop_cols)

    return train_res, test_res
```

```python
#3-5　FE(試走結果に基づくEDA(第1有力指標))　境界値フラグ(EXT_SOURCE_1～3)

def apply_ext_source_border_flags(df):

    # EXT_SOURCE_1～3の境界値フラグを作成
    # 境界値未満を '1'、境界値以上を '0' とする

    # 指定された境界値の設定
    borders = {
        'EXT_SOURCE_1': 0.4284,
        'EXT_SOURCE_2': 0.4895,
        'EXT_SOURCE_3': 0.4484
    }

    for col, threshold in borders.items():
        if col in df.columns:
            # 境界値未満なら1、以上なら0（欠損値は境界値以上ではないため0になります）
            df[f'{col}_border_frag'] = (df[col] < threshold).astype(int)

    return df
```

```python
#3-6　FE(試走結果に基づくEDA(第1有力指標))　業種別のデフォルト率(ORGANIZATION_TYPE)
from sklearn.model_selection import KFold

def apply_org_type_te(train_df, test_df, target_col='TARGET', n_folds=5, smoothing=10):
    train_res = train_df.copy()
    test_res = test_df.copy()

    col = 'ORGANIZATION_TYPE'
    new_col = 'ORG_TYPE_TARGET_RATE'

    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    global_mean = train_res[target_col].mean()

    # numpy配列に溜める
    oof_values = np.zeros(len(train_res), dtype=np.float64)

    for train_idx, val_idx in kf.split(train_res):
        df_t = train_res.iloc[train_idx]
        df_v = train_res.iloc[val_idx]

        agg = df_t.groupby(col)[target_col].agg(['count', 'mean'])
        counts = agg['count']
        means = agg['mean']
        smooth = (means * counts + global_mean * smoothing) / (counts + smoothing)

        # numpy配列に直接書き込み
        oof_values[val_idx] = df_v[col].map(smooth).values

    # 最後に一括代入
    train_res[new_col] = oof_values

    # Testデータ
    agg = train_res.groupby(col)[target_col].agg(['count', 'mean'])
    counts = agg['count']
    means = agg['mean']
    smooth = (means * counts + global_mean * smoothing) / (counts + smoothing)
    test_res[new_col] = test_res[col].map(smooth).values

    # oof_valuesの処理（forループの後）
    oof_values = np.where(np.isnan(oof_values), global_mean, oof_values)
    train_res[new_col] = oof_values

    # Testデータ
    agg = train_res.groupby(col)[target_col].agg(['count', 'mean'])
    counts = agg['count']
    means = agg['mean']
    smooth = (means * counts + global_mean * smoothing) / (counts + smoothing)
    test_values = test_res[col].map(smooth).values.astype(np.float64)
    test_values = np.where(np.isnan(test_values), global_mean, test_values)
    test_res[new_col] = test_values

    return train_res, test_res
```

```python
#4-1　FE(試走結果に基づくEDA(第2有力指標))　ビン(50)別デフォルト率

# 第2有力指標のリスト
secondary_importance_cols = [
    'AMT_ANNUITY',
    'INS_AMT_PAYMENT_min',
    'BU_DAYS_CREDIT_max',
    'AMT_CREDIT',
    'DAYS_BIRTH',
    'AMT_GOODS_PRICE'
]

from sklearn.model_selection import KFold

def apply_bin_target_encoding(train_df, test_df, target_cols, target_val='TARGET', n_bins=50, n_folds=5, smoothing=10):
    train_res = train_df.copy()
    test_res = test_df.copy()

    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    global_mean = train_res[target_val].mean()

    for col in target_cols:
        if col not in train_res.columns:
            continue

        new_col = f'{col}_BIN_TARGET_RATE'

        combined_min = train_res[col].min()
        combined_max = train_res[col].max()
        bins = np.linspace(combined_min, combined_max, n_bins + 1)

        train_res['tmp_bin'] = pd.cut(train_res[col], bins=bins, include_lowest=True, labels=False)
        test_res['tmp_bin'] = pd.cut(test_res[col], bins=bins, include_lowest=True, labels=False)

        oof_values = np.zeros(len(train_res), dtype=np.float64)

        for train_idx, val_idx in kf.split(train_res):
            df_t = train_res.iloc[train_idx]
            df_v = train_res.iloc[val_idx]

            agg = df_t.groupby('tmp_bin')[target_val].agg(['count', 'mean'])
            counts = agg['count']
            means = agg['mean']
            smooth = (means * counts + global_mean * smoothing) / (counts + smoothing)

            oof_values[val_idx] = df_v['tmp_bin'].map(smooth).values

        oof_values = np.where(np.isnan(oof_values), global_mean, oof_values)
        train_res[new_col] = oof_values

        agg = train_res.groupby('tmp_bin')[target_val].agg(['count', 'mean'])
        smooth = (agg['mean'] * agg['count'] + global_mean * smoothing) / (agg['count'] + smoothing)
        test_values = test_res['tmp_bin'].map(smooth).values.astype(np.float64)
        test_values = np.where(np.isnan(test_values), global_mean, test_values)
        test_res[new_col] = test_values

        train_res = train_res.drop(columns=['tmp_bin'])
        test_res = test_res.drop(columns=['tmp_bin'])

    return train_res, test_res
```

```python
#4-2　FE(試走結果に基づくEDA(第2有力指標))　キリ番(AMT_CREDIT,AMT_GOODS_PRICE)
def apply_round_number_flags(df):
    """
    AMT_CREDIT, AMT_GOODS_PRICE がキリの良い数字かどうかを判定
    1万, 10万, 100万で割り切れる（小数点以下がない）場合にフラグを立てる
    """
    cols = ['AMT_CREDIT', 'AMT_GOODS_PRICE']
    divisors = [10000, 100000, 1000000] # 1e4, 1e5, 1e6

    for col in cols:
        if col not in df.columns:
            continue

        for div in divisors:
            # 除算して余りが0（または非常に小さい誤差範囲）ならフラグ
            # divの大きさに合わせて 1e4なら "10k", 1e5なら "100k" のように命名
            label = f"{div/1000:g}k"
            new_col = f"{col}_is_round_{label}"

            # 欠損値は 0 (False) と判定されるように処理
            df[new_col] = (df[col] % div == 0).astype(int)

    return df
```

```python
#5-1　モデル試走(LGBM,CAT_K-FOLD)

import lightgbm as lgb
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score

def train_lgbm(X, y, X_test, params=None):
    """LightGBMによるK-Fold学習"""
    if params is None:
        params = {
            'objective': 'binary', 'metric': 'auc', 'verbosity': -1,
            'learning_rate': 0.05, 'num_leaves': 31, 'random_state': 42
        }

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    oof_preds = np.zeros(len(X))
    test_preds = np.zeros(len(X_test))

    for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        dtrain = lgb.Dataset(X_train, label=y_train)
        dval = lgb.Dataset(X_val, label=y_val, reference=dtrain)

        model = lgb.train(
            params, dtrain, valid_sets=[dtrain, dval],
            num_boost_round=10000,
            callbacks=[lgb.early_stopping(100), lgb.log_evaluation(0)]
        )

        oof_preds[val_idx] = model.predict(X_val)
        test_preds += model.predict(X_test) / kf.n_splits
        print(f"LGBM Fold {fold+1} AUC: {roc_auc_score(y_val, oof_preds[val_idx]):.5f}")

    print(f"LGBM Overall CV: {roc_auc_score(y, oof_preds):.5f}")
    return oof_preds, test_preds

def train_lgbm(X, y, X_test, params=None):
    if params is None:
        params = {'objective': 'binary', 'metric': 'auc', 'verbosity': -1, 'learning_rate': 0.05, 'num_leaves': 31, 'random_state': 42}
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    oof_preds = np.zeros(len(X))
    test_preds = np.zeros(len(X_test))
    models = [] # モデルを格納するリストを追加

    for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        dtrain = lgb.Dataset(X_train, label=y_train)
        dval = lgb.Dataset(X_val, label=y_val, reference=dtrain)

        model = lgb.train(params, dtrain, valid_sets=[dtrain, dval], num_boost_round=10000,
                          callbacks=[lgb.early_stopping(100), lgb.log_evaluation(0)])

        models.append(model) # リストに保存
        oof_preds[val_idx] = model.predict(X_val)
        test_preds += model.predict(X_test) / kf.n_splits
        print(f"LGBM Fold {fold+1} AUC: {roc_auc_score(y_val, oof_preds[val_idx]):.5f}")
    print(f"LGBM Overall CV: {roc_auc_score(y, oof_preds):.5f}")
    return oof_preds, test_preds, models # modelsを返す

def train_catboost(X, y, X_test, params=None):
    if params is None:
        params = {'loss_function': 'Logloss', 'eval_metric': 'AUC', 'learning_rate': 0.05, 'iterations': 10000, 'random_seed': 42, 'verbose': 0, 'early_stopping_rounds': 100}
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    oof_preds = np.zeros(len(X))
    test_preds = np.zeros(len(X_test))
    models = [] # モデルを格納するリストを追加

    for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

        model = CatBoostClassifier(**params)
        model.fit(X_train, y_train, eval_set=(X_val, y_val), use_best_model=True)

        models.append(model) # リストに保存
        oof_preds[val_idx] = model.predict_proba(X_val)[:, 1]
        test_preds += model.predict_proba(X_test)[:, 1] / kf.n_splits
        print(f"CatBoost Fold {fold+1} AUC: {roc_auc_score(y_val, oof_preds[val_idx]):.5f}")
    print(f"CatBoost Overall CV: {roc_auc_score(y, oof_preds):.5f}")
    return oof_preds, test_preds, models # modelsを返す
```

```python
#5-6　モデル寄与度

# --- 1. 寄与度の集計と正規化 (%) ---
def get_pct_importance(models, model_name, feature_names):
    if model_name == 'LGBM':
        # Gain（合計）を抽出して平均
        imp = np.mean([m.feature_importance(importance_type='gain') for m in models], axis=0)
    else:
        # CatBoostのImportanceを平均
        imp = np.mean([m.get_feature_importance() for m in models], axis=0)

    # 合計を100%に正規化
    imp_pct = (imp / np.sum(imp)) * 100
    return pd.DataFrame({'feature': feature_names, f'{model_name}_pct': imp_pct})

# 寄与度データ作成
lgb_imp_df = get_pct_importance(models_lgb_fe, 'LGBM', X.columns)
cat_imp_df = get_pct_importance(models_cat_fe, 'CAT', X.columns)

# --- 2. マージして比較表を作成 ---
comparison_df = pd.merge(lgb_imp_df, cat_imp_df, on='feature')

# AUCスコアの取得
lgb_auc = roc_auc_score(y, oof_lgb_fe)
cat_auc = roc_auc_score(y, oof_cat_fe)

# --- 3. 結果表示 ---
print(f" 精度比較 (AUC) :")
print(f"  - LightGBM: {lgb_auc:.4f}")
print(f"  - CatBoost: {cat_auc:.4f}\n")

# LGBMのTOP15
lgb_top15_s = comparison_df.sort_values(by='LGBM_pct', ascending=False).head(15)[['feature', 'LGBM_pct']]
lgb_top15_s.columns = ['LGBM Feature', 'Importance(%)']

# CatBoostのTOP15
cb_top15_s = comparison_df.sort_values(by='CAT_pct', ascending=False).head(15)[['feature', 'CAT_pct']]
cb_top15_s.columns = ['CatBoost Feature', 'Importance(%)']

# インデックスをリセットして横に並べる
lgb_top15_s = lgb_top15_s.reset_index(drop=True)
cb_top15_s = cb_top15_s.reset_index(drop=True)
side_by_side_s = pd.concat([lgb_top15_s, cb_top15_s], axis=1)

print(f" モデル別インポータンス TOP 15 ({len(X.columns)}列版)")
print(side_by_side_s.to_markdown(index=False))

# --- 共通 ---
common_top_s = set(lgb_top15_s['LGBM Feature']) & set(cb_top15_s['CatBoost Feature'])
print(f"\n 両モデルが共通してTOP15に挙げた項目: {len(common_top_s)}件")
print(sorted(list(common_top_s)))
```

```python
#5-7　追加カラムの寄与度

# --- 追加カラムに特化したランキング表示コード ---

def show_added_features_ranking(models_lgb, models_cat, feature_names):
    # 1. LGBMの寄与度(%)とランク
    lgb_imp = np.mean([m.feature_importance(importance_type='gain') for m in models_lgb], axis=0)
    lgb_pct = (lgb_imp / np.sum(lgb_imp)) * 100

    # 2. CatBoostの寄与度(%)とランク
    cat_imp = np.mean([m.get_feature_importance() for m in models_cat], axis=0)
    cat_pct = (cat_imp / np.sum(cat_imp)) * 100

    # データフレーム化
    df = pd.DataFrame({
        'feature': feature_names,
        'LGBM_pct': lgb_pct,
        'CAT_pct': cat_pct
    })

    # 全体での順位を計算
    df['LGBM_Rank'] = df['LGBM_pct'].rank(ascending=False).astype(int)
    df['CAT_Rank'] = df['CAT_pct'].rank(ascending=False).astype(int)

    # 追加カラムのキーワードでフィルタリング
    added_keys = ['abnormal_MAX_frag', '_TE', '_mean', 'border', 'round', 'geometric']
    added_df = df[df['feature'].str.contains('|'.join(added_keywords))].copy()

    # 表示用に整形（CATの順位が高い順に並べる）
    added_df = added_df.sort_values(by='CAT_Rank')

    print(f"=== 追加カラム限定：寄与度(%) & 全体順位 (全{len(feature_names)}列中) ===")

    # カラム名をわかりやすくリネーム
    result = added_df[[
        'feature', 'LGBM_pct', 'LGBM_Rank', 'CAT_pct', 'CAT_Rank'
    ]].reset_index(drop=True)

    # 表形式で表示
    print(result.to_markdown(index=False))

    # 簡易サマリー
    top100_lgb = len(result[result['LGBM_Rank'] <= 100])
    top100_cat = len(result[result['CAT_Rank'] <= 100])
    print(f"\n[Summary]")
    print(f"TOP 100入りした追加カラム: LGBM {top100_lgb}個 / CatBoost {top100_cat}個")

# 実行
show_added_features_ranking(models_lgb_fe, models_cat_fe, X.columns)
```
