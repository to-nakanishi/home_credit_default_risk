```python
#1-4 基本EDA (異常値)

# 数値型カラムのみ対象
num_df = train.select_dtypes(include=['number'])

# 1. 統計量の計算（Q95とMax）
Q95 = num_df.quantile(0.95)
Max = num_df.max()
Mean = num_df.mean()
Q50 = num_df.quantile(0.50)

# 2. Max / Q95 比率を計算
# Q95が0の場合、除算エラーを避けるため微小値を加算
ratios = Max / (Q95 + 1e-9)

# 3. サマリーの作成
outlier_summary = pd.DataFrame({
    'Max/Q95比': ratios,
    '最大値': Max,
    '95%点': Q95,
    '中央値': Q50,
    '平均': Mean
})

# --- 判定基準 ---
# 比率が 100倍 を超えるものを「不自然な跳ね上がり」として抽出
# かつ、最大値、95％値が0でないもの
condition = (outlier_summary['Max/Q95比'] > 100) & (outlier_summary['最大値'] != 0) & (outlier_summary['95%点'] != 0)
outlier_summary = outlier_summary[condition].sort_values('Max/Q95比', ascending=False)

print(f"不自然な跳ね上がりが見つかったカラム数: {len(outlier_summary)} / {len(num_df.columns)}")
display(outlier_summary)
```

```python
#2-1　ターゲット変数(X日)　installments_paymentsへのターゲット変数結合
'''
以下のとおり、返済の遅延開始と遅延日数についてはマスクされているため確認を行う
　「目的変数（1 - 支払い困難を抱える顧客：サンプル中のローンの最初のY回の分割払いのうち、
　　少なくとも1回でX日以上支払いが遅れた顧客、0 - その他のすべてのケース）」

以下の算式をもって検討
X日：installments_payments　（DAYS_ENTRY_PAYMENT - DAYS_INSTALMENT）　支払日-期限日
'''
# 1. train から必要な列（キーとターゲット）だけを抽出
target_map = train[['SK_ID_CURR', 'TARGET']]

# 2. instalments_payments に TARGET を結合（左結合：left join）
# ※ instalments_payments の全行を残しつつ、顧客IDが一致する TARGET を紐づけ
instalments_payments = instalments_payments.merge(target_map, on='SK_ID_CURR', how='left')

# 3. ターゲットが紐づかない行（testデータに含まれる顧客など）をどうするか確認
print(f"結合後の行数: {len(instalments_payments)}")
print(f"TARGETが欠損している行数: {instalments_payments['TARGET'].isnull().sum()}")

# 分析のために、TARGETが判明している（trainに含まれる）データだけに絞り込む
ins_train = instalments_payments.dropna(subset=['TARGET']).copy()
ins_train['TARGET'] = ins_train['TARGET'].astype(int) # floatをintに直す

print(f"分析対象の行数: {len(ins_train)}")
```

```python
#2-2　基本EDA　ターゲット変数(X日)　遅延日数(DPD)ヒストグラム
# 1. 支払遅延日数(DPD)を計算
ins_train['DPD'] = ins_train['DAYS_ENTRY_PAYMENT'] - ins_train['DAYS_INSTALMENT']

# 2. 顧客ごとに最大値を抽出（一旦変数に代入）
user_max_dpd = ins_train.groupby('SK_ID_CURR')['DPD'].max().reset_index()

# 3. TARGETと結合
analysis_raw_df = user_max_dpd.merge(train[['SK_ID_CURR', 'TARGET']], on='SK_ID_CURR', how='left')

# 4. 可視化
plt.figure(figsize=(6, 4))
# 変数名を DPD に統一
sns.histplot(data=analysis_raw_df[(analysis_raw_df['DPD'] >= -50) & (analysis_raw_df['DPD'] <= 120)],
             x='DPD', hue='TARGET',
             bins=85, element='step', common_norm=False, stat='density', alpha=0.5)

# 0日のライン（遅延の境界）
plt.axvline(x=0, color='black', linestyle='-', alpha=0.5, label='On-time (0)')

plt.title('Distribution of Maximum Delay Days per Customer (Raw Max DPD)')
plt.xlabel('Max DPD (Negative values indicate early payment)')
plt.ylabel('Density')
plt.legend(title='TARGET')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
#2-3　ターゲット変数(X日)　DPD統計量

# 1. ターゲットを紐付けた一時的なデータフレームを作成
# メモリ節約のため必要な列だけ抽出
cols = ['DAYS_ENTRY_PAYMENT', 'DAYS_INSTALMENT', 'DPD']
tmp = ins_train[cols + ['SK_ID_CURR']].merge(train[['SK_ID_CURR', 'TARGET']], on='SK_ID_CURR', how='left')

# 2. カラムごとに TARGET別の統計量を算出
analysis_results = []

for col in cols:
    # TARGETごとに 件数(count:非欠損数) と 欠損率 を計算
    res = tmp.groupby('TARGET')[col].agg([
        ('Valid_Count', 'count'),
        ('Missing_Rate', lambda x: x.isnull().mean())
    ])
    res['Column'] = col
    analysis_results.append(res)

# 結合して見やすく表示
df_null_analysis = pd.concat(analysis_results).reset_index()
df_null_analysis = df_null_analysis[['Column', 'TARGET', 'Valid_Count', 'Missing_Rate']]

print("=== Column-wise Analysis by TARGET ===")
print(df_null_analysis.to_string(index=False))
```

```python
#3-1 試走結果に基づくEDA(第1有力指標)　欠損及び基本統計量
'''
第1有力指標 (5%以上)
共通： EXT_SOURCE_1～3 (外部機関によるスコア)
LGBM： ORGANIZATION_TYPE (勤務先の業種)
'''
core_cols1 = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']

# 統計量と欠損率
def get_stats_summary(df, cols):
    """
    指定したカラムの欠損率と基本統計量をDataFrameで返す
    """
    summary = pd.DataFrame({
        'Missing%': df[cols].isnull().mean() * 100,
        'Mean': df[cols].mean(numeric_only=True),
        'Median': df[cols].median(numeric_only=True),
        'Min': df[cols].min(numeric_only=True),
        'Max': df[cols].max(numeric_only=True)
    })
    return summary

core_summary = get_stats_summary(train, core_cols1)
print(core_summary)
```

```python
#3-4　EXT_SOURCE_1〜3 × TARGET ヒストグラム(密度表示)の境界値

from sklearn.neighbors import KernelDensity

def find_intersection(data, col):
    # 欠損を除外
    mask = data[col].notnull()
    s0 = data.loc[mask & (data['TARGET'] == 0), col].values
    s1 = data.loc[mask & (data['TARGET'] == 1), col].values

    # 各グループの密度推定（KDE）を計算
    # スパン（x軸）を0から1まで1000分割
    x_range = np.linspace(0, 1, 1000).reshape(-1, 1)

    # TARGET=0の密度
    kde0 = KernelDensity(bandwidth=0.03).fit(s0.reshape(-1, 1))
    log_dens0 = kde0.score_samples(x_range)

    # TARGET=1の密度
    kde1 = KernelDensity(bandwidth=0.03).fit(s1.reshape(-1, 1))
    log_dens1 = kde1.score_samples(x_range)

    # 密度が逆転するインデックスを探す
    diff = np.exp(log_dens0) - np.exp(log_dens1)
    idx = np.argwhere(np.diff(np.sign(diff))).flatten()

    if len(idx) > 0:
        return x_range[idx[-1]][0] # 最も右側（高スコア側）の交点を返す
    return None

# 実行と表示
print("--- EXT_SOURCE Border Lines ---")
for col in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']:
    border = find_intersection(train, col)
    print(f"{col}: {border:.4f}")
```

```python
#3-4　試走結果に基づくEDA(第1有力指標)　ORGANIZATION_TYPE

# 1. 業種ごとのターゲット平均（デフォルト率）を計算
org_target_mean = train.groupby('ORGANIZATION_TYPE')['TARGET'].mean().sort_values(ascending=False)

# 2. 可視化
plt.figure(figsize=(10, 18)) # 種類が多いので縦に長く
sns.barplot(x=org_target_mean.values, y=org_target_mean.index, color='steelblue', edgecolor='black')

# 全体の平均デフォルト率を赤い破線で引く（比較用）
overall_mean = train['TARGET'].mean()
plt.axvline(overall_mean, color='red', linestyle='--', label=f'Overall Average ({overall_mean:.2%})')

plt.title('Default Rate by Organization Type')
plt.xlabel('Default Rate (TARGET Mean)')
plt.legend()
plt.show()
```

```python
#3-5　試走結果に基づくEDA(第1有力指標)　ORGANIZATION_TYPE カイ二乗検定

from scipy import stats

# 1. 業種(ORGANIZATION_TYPE)とターゲット(TARGET)のクロス集計表を作成
# 行：業種、列：0(完済) or 1(デフォルト)
contingency_table = pd.crosstab(train['ORGANIZATION_TYPE'], train['TARGET'])

# 2. カイ二乗検定を実行
# chi2: 検定統計量, p: p値, dof: 自由度, expected: 期待度数
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"--- ORGANIZATION_TYPE Chi-square Test ---")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4e}") # 非常に小さい値になるため科学的表記

# 3. 業種別のデフォルト率を算出して、上位を確認
org_stats = train.groupby('ORGANIZATION_TYPE')['TARGET'].agg(['mean', 'count'])
org_stats.columns = ['Default_Rate', 'Count']
org_stats = org_stats.sort_values(by='Default_Rate', ascending=False)

print("\n--- Default Rate by Organization (Top 5) ---")
print(org_stats.head())
```

P値が限りなく小さく、有意な差であることを確認
