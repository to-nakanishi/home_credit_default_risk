```python
#5-1 返済能力(借入、収入、支出)　欠損及び基本統計量
'''
・借入関連カラム
AMT_CREDIT(融資額)、AMT_ANNUITY (ローンの年間支払い額)、AMT_GOODS_PRICE(商品価格)
・収入関連カラム
AMT_INCOME_TOTAL(顧客の収入)
・支出関連カラム
FLAG_OWN_CAR(車所有フラグ)、CNT_CHILDREN(子供の数)、NAME_FAMILY_STATUS(家族構成)
NAME_HOUSING_TYPE(住宅状況(賃貸等))、REGION_POPULATION_RELATIVE(居住地の人口密度)
CNT_FAM_MEMBERS (世帯人数)
bureau.csv　AMT_CREDIT_SUM_DEBT(信用情報機関の信用情報に記載されている現在の債務)
※BU_AMT_CREDIT_SUM_DEBT_latest
POS_CASH_balance.csv　CNT_INSTALMENT_FUTURE(前回のクレジットの残りの支払い額)
※POS_CNT_INSTALMENT_FUTURE_latest
'''

def get_stats_summary(df, cols):
    """
    指定したカラムの欠損率と基本統計量をDataFrameで返す
    """
    summary = pd.DataFrame({
        'Missing%': df[cols].isnull().mean() * 100,
        'Mean': df[cols].mean(numeric_only=True),
        'Median': df[cols].median(numeric_only=True),
        'Min': df[cols].min(numeric_only=True),
        'Max': df[cols].max(numeric_only=True)
    })
    return summary

repayment_capacity_cols1 = [
    # --- 借入関連 ---
    'AMT_CREDIT',          # 融資額
    'AMT_ANNUITY',         # ローンの年間支払い額
    'AMT_GOODS_PRICE',     # 商品価格

    # --- 収入関連 ---
    'AMT_INCOME_TOTAL',    # 顧客の収入

    # --- 支出・属性関連 ---
    'FLAG_OWN_CAR',        # 車所有フラグ
    'CNT_CHILDREN',        # 子供の数
    'NAME_FAMILY_STATUS',  # 家族構成
    'NAME_HOUSING_TYPE',   # 住宅状況
    'REGION_POPULATION_RELATIVE', # 居住地の人口密度
    'CNT_FAM_MEMBERS',     # 世帯人数

    # --- 他社債務・支払い状況（集約済み最新値） ---
    'BU_AMT_CREDIT_SUM_DEBT_latest',   # 信用情報機関の債務残高
    'POS_CNT_INSTALMENT_FUTURE_latest' # POS契約の残り支払い回数
]
repayment_capacity_summary = get_stats_summary(train, repayment_capacity_cols1)
print(repayment_capacity_summary)
```

```python
#5-2 返済能力(借入、収入、支出)　カテゴリ型の基本情報

def analyze_category_risk(df, cols, target_col='TARGET'):
    for col in cols:
        if col in df.columns:
            print(f"\n▼ カラム名: {col}")
            print("-" * 80)

            # 1. 基本集計（件数、構成比、デフォルト率）
            summary = df.groupby(col, dropna=False).agg(
                件数=(target_col, 'size'),
                デフォルト件数=(target_col, 'sum'),
                デフォルト率=(target_col, 'mean')
            )

            # 構成比(%)を追加
            summary['構成比(%)'] = (summary['件数'] / len(df)) * 100
            # デフォルト率を%表記に
            summary['デフォルト率(%)'] = summary['デフォルト率'] * 100

            # 2. 欠損率の算出（全体に対して）
            missing_pct = df[col].isnull().mean() * 100

            # 表示用に整形（ユニーク数の降順を意識しつつ、デフォルト率でソート）
            res = summary[['件数', '構成比(%)', 'デフォルト率(%)']].sort_values('デフォルト率(%)', ascending=False)

            print(res)
            print(f"\n[補足] このカラムの全体欠損率: {missing_pct:.2f}%")
            print("-" * 80)


repayment_capacity_cols1 = ['NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'FLAG_OWN_CAR']
analyze_category_risk(train, repayment_capacity_cols1)
```

```python
#7-1　申告情報の整合性　欠損及び欠損及び基本統計量
'''
AMT_GOODS_PRICE(商品価格)-AMT_CREDIT(融資額)
DAYS_ID_PUBLISH(申請時の身分証更新日)
REG_REGION_NOT_LIVE_REGION(連絡先と居住地の相違フラグ)
FLAG_MOBIL(携帯番号の提供有無のフラグ)
FLAG_EMP_PHONE(勤務先の電話番号の提供有無のフラグ)
FLAG_WORK_PHONE(自宅の電話番号の提供有無のフラグ)
FLAG_CONT_MOBILE(携帯番号と連絡可否のフラグ)
FLAG_EMAIL(メールアドレスの提供有無があるかどうかのフラグ)
bureau.csv　CREDIT_DAY_OVERDUE(サンプルにおける関連ローン申請時点でのCBクレジットの延滞日数)
※BU_CREDIT_DAY_OVERDUE_max,BU_CREDIT_DAY_OVERDUE_latest
previous_application.csv　CODE_REJECT_REASON(前回の申請が却下された理由)
※PREV_CODE_REJECT_REASON_latest
'''
train['DIFF_GOODS_CREDIT'] = train['AMT_GOODS_PRICE'] - train['AMT_CREDIT']
false_features1 = [
    'DIFF_GOODS_CREDIT',           # AMT_GOODS_PRICE - AMT_CREDIT
    'DAYS_ID_PUBLISH',             # 申請時の身分証更新日
    'BU_CREDIT_DAY_OVERDUE_max',   # 他社借入の最大延滞日数
    'BU_CREDIT_DAY_OVERDUE_latest',# 他社借入の直近延滞日数
]

false_features1_summary = get_stats_summary(train,false_features1)
print(false_features1_summary)
```

```python
#8-1　その他　(同行者、年齢)
# NAME_TYPE_SUITE × age_group でデフォルト率を算出

df_plot = train.copy()

if 'age_group' not in df_plot.columns:
    # DAYS_BIRTHはマイナス表記なので、正の数にして365で割り年齢を出す
    df_plot['age'] = df_plot['DAYS_BIRTH'] / -365
    df_plot['age_group'] = (df_plot['age'] // 10 * 10).astype(int).astype(str) + 's'

col_x = 'NAME_TYPE_SUITE'
col_y = 'age_group'

res_suite = df_plot.pivot_table(
    index=col_x,
    columns=col_y,
    values='TARGET',
    aggfunc='mean'
) * 100

print(f"▼ {col_x} × 年齢層別 デフォルト率(%)")
print("-" * 80)
print(res_suite.round(2))
```

```python
#8-2　その他　(同行者、直前1時間の照会有無)
# 欠損値がある場合、集計が偏るため 0 で埋める（またはそのままでも可）
df_plot['REQ_HOUR_CLIP'] = df_plot['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(0)

# 直前1時間の照会がある(1回以上)か、ない(0回)かでフラグ化
df_plot['IS_URGENT_REQ'] = df_plot['REQ_HOUR_CLIP'].apply(lambda x: 1 if x > 0 else 0)

# NAME_TYPE_SUITE × IS_URGENT_REQ でデフォルト率を算出
res_urgent = df_plot.pivot_table(
    index='NAME_TYPE_SUITE',
    columns='IS_URGENT_REQ',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示を整える
res_urgent.columns = ['件数(照会なし)', '件数(あり)', 'デフォルト率(なし)%', 'デフォルト率(あり)%']
res_urgent['デフォルト率(なし)%'] *= 100
res_urgent['デフォルト率(あり)%'] *= 100

print(f"▼ 同行者 × 直前1時間の照会有無 リスクマトリックス")
print("-" * 85)
print(res_urgent.round(2).sort_values('デフォルト率(あり)%', ascending=False))
print("-" * 85)
```

子連れや他人Aタイプの同行者は直前１時間の照会があるとデフォルト率が高い傾向あり。

```python
#8-3　その他　(同行者、当日(24h内)の照会有無)
# 欠損値を0で埋め、当日の照会有無をフラグ化
df_plot['REQ_DAY_CLIP'] = df_plot['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(0)
df_plot['IS_DAY_REQ'] = df_plot['REQ_DAY_CLIP'].apply(lambda x: 1 if x > 0 else 0)

# NAME_TYPE_SUITE × IS_DAY_REQ でデフォルト率を算出
res_day = df_plot.pivot_table(
    index='NAME_TYPE_SUITE',
    columns='IS_DAY_REQ',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示整理
res_day.columns = ['件数(なし)', '件数(あり)', 'デフォルト率(なし)%', 'デフォルト率(あり)%']
res_day['デフォルト率(なし)%'] *= 100
res_day['デフォルト率(あり)%'] *= 100

print(f"▼ 同行者 × 当日(24h内)の照会有無 リスクマトリックス")
print("-" * 85)
print(res_day.round(2).sort_values('デフォルト率(あり)%', ascending=False))
print("-" * 85)
```

当日中に照会がある場合の家族連れや子連れのデフォルト率は高い。

```python
#8-4　その他　(住宅形態、人口密度別)
# 子供の数を 0, 1, 2, 3以上 の4区分に整理（外れ値の影響を防ぐ）
# 人口密度を5段階に分割
df_plot['pop_group'] = pd.qcut(df_plot['REGION_POPULATION_RELATIVE'], 5, labels=['過疎', 'やや低', '中間', 'やや高', '密集'])

# 住宅形態 × 人口密度別 デフォルト率(%)
res_pop_house = df_plot.pivot_table(
    index='NAME_HOUSING_TYPE',
    columns='pop_group',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示整理
res_pop_house_mean = res_pop_house['mean'] * 100

print(f"▼ 住宅形態 × 人口密度別 デフォルト率(%)")
print("-" * 90)
print(res_pop_house_mean.round(2))
print("-" * 90)
```

想定とは異なり、密集地域でのデフォルト率は低く、中間地域が概ね高い傾向にある。また密度に比例せず、同一密度であっても住宅形態によって大きく差が出る結果となった。

```python
#8-5　その他　(住宅形態、地域都市乗算値(1-9)別 デフォルト率(%))
# 1. 欠損値をそれぞれの平均値で補完
df_plot['REGION_RATING_CLIENT_fixed'] = df_plot['REGION_RATING_CLIENT'].fillna(df_plot['REGION_RATING_CLIENT'].mean())
df_plot['REGION_RATING_CLIENT_W_CITY_fixed'] = df_plot['REGION_RATING_CLIENT_W_CITY'].fillna(df_plot['REGION_RATING_CLIENT_W_CITY'].mean())

# 2. 乗算値の算出（1〜9の範囲になる）
df_plot['REGION_RATING_MUL'] = df_plot['REGION_RATING_CLIENT_fixed'] * df_plot['REGION_RATING_CLIENT_W_CITY_fixed']

# 3. 乗算値ごとのデフォルト率を集計
res_mul = df_plot.groupby('REGION_RATING_MUL')['TARGET'].agg(['count', 'mean'])
res_mul['mean'] *= 100

print("▼ 地域評価×都市評価（乗算値）別のデフォルト率(%)")
print("-" * 60)
print(res_mul.round(2).sort_index())
print("-" * 60)

# 乗算値（REGION_RATING_MUL）を整数型にして見やすくする（平均補完した場合は四捨五入）
df_plot['MUL_INT'] = df_plot['REGION_RATING_MUL'].round().astype(int)

# 住宅形態 × 乗算値（総合評価スコア）別 デフォルト率(%)
res_final = df_plot.pivot_table(
    index='NAME_HOUSING_TYPE',
    columns='MUL_INT',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示整理
res_final_mean = res_final['mean'] * 100

print(f"▼ 住宅形態 × 地域都市乗算値(1-9)別 デフォルト率(%)")
print("-" * 110)
print(res_final_mean.round(2))
print("-" * 110)
```

地域評価の乗算値においても単純な相関ではなく、3と6は高いが4は低くなる結果となった。

```python
#8-6　その他　(組織形態、申請時間帯別 デフォルト率(%))
# 24時間を「早朝・深夜」「日中(勤務中)」「夕方・夜」に分ける
def get_time_slot(h):
    if 9 <= h <= 17: return '1:日中(9-17)'
    if 18 <= h <= 23: return '2:夜間(18-23)'
    return '0:早朝深夜(0-8)'

df_plot['time_slot'] = df_plot['HOUR_APPR_PROCESS_START'].apply(get_time_slot)

# 組織種別 × 時間帯別 デフォルト率(%)
res_org_time = df_plot.pivot_table(
    index='ORGANIZATION_TYPE',
    columns='time_slot',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示整理（デフォルト率が高い順に並べる）
res_org_time_mean = res_org_time['mean'] * 100
print(f"▼ 組織形態 × 申請時間帯別 デフォルト率(%)")
print("-" * 100)
print(res_org_time_mean.round(2).sort_values('1:日中(9-17)', ascending=False).head(20))
print("-" * 100)
```

概ね、早朝深夜にかけてはデフォルト率が高いが業種によっては低くなる場合もある。

```python
#8-7　その他　(組織形態、申請時間帯別 デフォルト率(%))
# 職種 × 3時間帯別 デフォルト率(%)
res_occ_time = df_plot.pivot_table(
    index='OCCUPATION_TYPE',
    columns='time_slot',
    values='TARGET',
    aggfunc=['count', 'mean']
)

# 表示整理
res_occ_mean = res_occ_time['mean'] * 100

print(f"▼ 職種 × 申請時間帯別 デフォルト率(%)")
print("-" * 100)
# 全体のデフォルト率が高い順に並べて表示
print(res_occ_mean.round(2).sort_values('1:日中(9-17)', ascending=False))
print("-" * 100)
```

業種別では夜間は比較的デフォルト率が低い傾向があったが、職種では夜間の方が高くなるケースも見られた。
